```python
import matplotlib.pyplot as plt

# X-axis values
sizes = [1, 10, 100]

# Raw CSV times
raw_times = [0.0102, 0.01804, 0.02098]
# Normalized DB (only measured at 100 MB)
norm_sizes = [100]
norm_times = [0.03854]
plt.figure(figsize=(8,5))
# Raw CSV line
plt.plot(sizes, raw_times, marker='o', label='Raw CSV')
# Normalized DB point
plt.scatter(norm_sizes, norm_times, color='orange', s=100,
            label='Normalized Database (100 MB)')

plt.xlabel("Data Size (MB)")
plt.ylabel("Query Time (seconds)")
plt.title("Query Performance vs Data Size")
plt.xticks([1, 10, 100])  # IMPORTANT
plt.legend()
plt.grid(True)

plt.show()
```
